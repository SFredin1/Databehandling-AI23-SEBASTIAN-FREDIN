# Exercises: Filter and query data

```python
import pandas as pd
autos = pd.read_json("../Data/new_autos.json")
```

### Use the "autos" dataset to answer the following questions:

Q1: How many cars from Jaguar are in the dataset, and what is the price of the most expensive one?

Answer
- count=3    
- price=36000.0

```python
autos.query("make == 'jaguar'")
```

### Q2: How many cars from Toyota are in the dataset, and what is the price of the most expensive one?

Answer:

- count=32    
- price=17669.0

```python
autos.query("make == 'toyota'")
```

### Q3: What is the length, width and height of the most expensive car in the entire dataset?

Answer:
- length=199.2    
- width=72.0    
- height=55.4

```python
autos.sort_values(by=["price"], ascending=[False])
autos.sort_values(by=["width"], ascending=[False])
autos.sort_values(by=["height"], ascending=[False])
```

```python
autos.sort_values("price")
```

### Q4: What is the lowest price per horsepower in the dataset, and what brand ("make") is that car?

Answer:
- price per horsepower=72.84    
- brand=Toyota

```python
autos.sort_values(by=["horsepower", "price"])
```
