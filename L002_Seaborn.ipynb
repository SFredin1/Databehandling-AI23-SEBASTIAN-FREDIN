# Seaborn
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip" ,hue="sex", style="smoker", size="size", col="day", row="time", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip" ,hue="sex", style="smoker", size="size", col="day", col_wrap=2, data=tips)
```

```python
sns.relplot(x="total_bill", y="tip" ,hue="sex", style="smoker", size="size", sizes=(20, 100) ,data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="bar", data=tips)
```

```python
sns.displot(data=tips, x="day", y="total_bill")
```
