## Combining datasets
Pandas provides various facilities for easily comining together Series or Dataframes.
- **Concat:** combining DataFrames across rows or columns.
- **Join:** Comining data on a key column or an index.
- **Merge:** combining data on common columns or indexes. 

```python
from IPython.display import display_html
import pandas as pd
from helpers import sample_df, hdisplay, nowrap_display
```

### Concat
Pandas .concat() method concatinates dataframe row-, or columnwise,<br>
with optional ser logic (union or intersection) of the indexes on the other axis.

Concat simply stacks multiple dataframes together either vertically, or horizontally after alligning on row/column.

```python
# Sample data
left = sample_df("A0", "D3")
right = sample_df("A0", "D3")

hdisplay([left, right], ["Left", "Right"])
```

```python
hdisplay([
    pd.concat([left, right], axis="index"),
    pd.concat([left, right], axis="columns")],
    ["axis='index'", "axis='columns'"]
)
```

```python
pd.concat([left, right]).reset_index(drop=True)
pd.concat([left, right]).set_index("C")

hdisplay([
    pd.concat([left, right], ignore_index=True),
    pd.concat([left, right], ignore_index=True),
    pd.concat([left, right], ignore_index=True),
])
```

```python
# New sample data

left = sample_df("A0", "D3", prefix="L_")
right = sample_df("C2", "F5", prefix="R_")

hdisplay([left, right], ["Left", "Right"])
```

```python
hdisplay([
    pd.concat([left, right], axis='index'),
    pd.concat([left, right], axis='columns', join='outer'), # union (default)
    pd.concat([left, right], axis='columns', join='inner')], # intersection
    ["axis='index'", "axis='columns', join='outer'", "axis='columns', join='inner'"],
    20
    )
```

```python
hdisplay([
    pd.concat([left, right], axis='index', keys=['left', 'right']), # multilevel index
    pd.concat([left, right], axis='columns', keys=['left', 'right']).loc[:, 'right']],
    ["axis='index', keys=['left', 'right']", "axis='columns', keys=['left', 'right']"]
    ) 
```

## Join
Join first alligns the index of two dataframes, and then picks up the remaining columns from the alligned rows of each dataframe.

The **how** parameter can be any of the following:
- **left:** Get *all* rows from the left table, and join matching rows from the right table. **(default)**
- **right:** Get *all* rows from the right table, and join matching rows from the left table.
- **outer:** Get *all* rows from both tables, joining all matching rows on both sides. **(union)**
- **inner:** Get only rows that exist in both tables. **(intersection)**
- **cross:** Get every possible comination of rows from both tables. Length of new table == len(left)*len(right) 

```python
# New sample data
left = sample_df("A0", "D3", prefix="L").add_prefix("L")
right = sample_df("C2", "F5", prefix="R").add_prefix("R")
hdisplay([left, right], ["Left", "Right"])
```

```python
hdisplay([
    left.join(right, how="left"),
    left.join(right, how="right")],
    ["how='left'", "how='right'"]
    )
```

```python
hdisplay([
    left.join(right, how="inner"),
    left.join(right, how="outer")],
    ["how='inner'", "how='outer'"]
    )
```

```python
left.join(right, how="cross")
```

## Merge
Pandas .merge() method is similar to join(), but is more versitile,
and allows us to specify columns beside the index to join on for both dataframes.

Note: *how* parameter works therefor the same as for join(), but the default for merge() is "inner", not "left"

```python
# New sample data
left = sample_df("A0", "E3").add_prefix("L")
left.loc[:, "LF"] = ["F8", "F9", "F10", "F11"]
right = sample_df("F10", "J13").add_prefix("R")
hdisplay([left, right], ["Left", "Right"])
```

```python
left.merge(right, how='right', left_on='LF', right_on='RF')
```
